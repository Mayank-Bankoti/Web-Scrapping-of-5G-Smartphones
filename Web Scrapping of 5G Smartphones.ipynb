### Part 2 - Pagination 

```python
Products= []             #List to store the name of the product
Prices= []               #List to store price of the product
RAM_ROM = []             #List to store RAM_ROM of the product
Display = []             #List to store Display of the product             
Camera = []              #List to store Camera of the product
Battery = []             #List to store Battery of the product
Processor = []           #List to store Processor of the product
ratings = []             #List to store ratings of the product
```

### Extract data from 10 pages using loop

```python
for i in range (1,11):

    # website in variable
    website = 'https://www.flipkart.com/search?q=Mobiles+5g&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(2)

    # request to website
    response = requests.get(website)

    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')

    for data in soup.findAll('div',class_='_2kHMtA'):
        names=data.find('div', attrs={'class':'_4rR01T'})
        price=data.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating=data.find('div', attrs={'class':'_3LWZlK'})
        specification = data.find('div', attrs={'class':'fMghEO'})

        for each in specification:
            col =each.find_all('li', attrs={'class':'rgWa7D'})
            ram_rom =col[0].text
            display = col[1].text
            camera = col[2].text
            battery = col[3].text
            processor = col[4].text
            Products.append(names.text)   # Add product name to list
            ratings.append(rating.text)   #Add rating specifications to list
            Battery.append(battery)       # Add Battery to list
            Prices.append(price.text)     # Add price to list
            Camera.append(camera)         #Add Camera specifications to list
            RAM_ROM.append(ram_rom)       #Add RAM_ROM specifications to list
            Display.append(display)       # Add Display specifications to list
            Processor.append(processor)   # Add Processor specifications to list

# After executing the above code, all the features are extracted and stored in the respective lists.

#printing the length of list
print(len(Products))
print(len(ratings))
print(len(Camera))
print(len(Prices))
print(len(RAM_ROM))
print(len(Display))
print(len(Battery))
print(len(Processor))
```

### Create Pandas Dataframe

```python
df=pd.DataFrame({'Product Name':Products,'Rating':ratings,'RAM_ROM':RAM_ROM,'Display':Display,'Camera':Camera,'Battery':Battery,'Processor':Processor,'Price':Prices})
```
